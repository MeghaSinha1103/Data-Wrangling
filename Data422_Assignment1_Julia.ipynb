### Question 1 (10 marks)

In which season would say we see more Little penguins?

_hint_ their scientific name is _Eudyptula minor_ and their Species abbreviation is "EUDMIN".

_hint_ you may want to use the package `Query.jl`.

### Answer ->
Most Little Penguins are seen in the winter season as shown by the code below.


```julia

#import Pkg
#Pkg.add("Query")
using Query, DataFrames
```

```julia
# Dropping the missing values in the COUNT column

new_df=dropmissing!(JoinData,:COUNT)


```

```julia
# Filtering the joined table based on the speciesAbbreviation as EUDMIN and grouping by season
new_df = JoinData|>
  @filter(_.SpeciesAbbreviation == "EUDMIN") |>
  @groupby(_.SEASN) |>
  @map({SeasonKey = key(_),COUNT = sum(_.COUNT) })

```

```julia
# Sorting the above table based on the COUNT. We get that the maximum value is for the Season winter.
x = @from i in new_df begin
    @orderby descending(i.COUNT), i.SeasonKey
    @select i
    @collect DataFrame
end

y = x |>
 @filter(_.COUNT == maximum(x[:COUNT]))

print(y)
 
```
