### <span style="color:black">Extracting All the links</span>

```julia
Alllinks = getattr.(c2_part_BaseLink,"href")
```


#### <span style="color:blue">Declaring function gor getting links</span>

```julia
function get_link(link)
  try
    response = HTTP.request("link") # we read the right page glueing together the base url and the link we got before 
    pagebad = response.body |> String |> parsehtml
    result_string = nodeText(pagebad.root)
    df2 = DataFrame(Link = link)
    return df2
  catch
    result_string = "no html content" 
    df2 = DataFrame(Link = link)
    return df2
  end
end
```

#### <span style="color:blue">Mapping links</span>

```julia
Links = vcat(map(get_link,Alllinks)...)
```

#### <span style="color:blue">Binding it with the previous dataframe</span>

```julia
df2 = hcat(df1,Links,makeunique=true) 
```

### <span style="color:black">Function for Automating the process of Scrapping different columns</span>

```julia
## Automating the process in Julia
function get_Data(genre)
    Baselink = "https://www.goodreads.com/giveaway/genre/"
    response = HTTP.request("GET","$Baselink$genre")
    parsed_page = response.body |> String |> parsehtml
    sel_for_c2 = Selector(".bookTitle")
    c2_part_BaseLink = eachmatch(sel_for_c2,parsed_page.root)
    bookTitle = []
    for i = 1:length(c2_part_BaseLink)
        value = c2_part_BaseLink[i][1].text
        push!(bookTitle, value)
   
    end
    df1 = DataFrame(Title = bookTitle)
    Alllinks = getattr.(c2_part_BaseLink,"href")
    Links = vcat(map(get_link,Alllinks)...)
    df2 = hcat(df1,Links,makeunique=true) 
    AuthorName = vcat(map(get_AuthorName,Alllinks)...)
    df3 = hcat(df2,AuthorName,makeunique=true) 
    Ratings = vcat(map(get_Rating,Alllinks)...)
    df4 = hcat(df3,Ratings,makeunique=true) 
    Description = vcat(map(get_Description,Alllinks)...)
    df5 = hcat(df4,Description,makeunique=true) 
    return(df5)
end
```
